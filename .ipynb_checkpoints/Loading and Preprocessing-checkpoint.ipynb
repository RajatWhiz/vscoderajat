```python
files= os.listdir('/s3data')
```

```python
def getNumericalAndCategorical(data):
    #get Numerical variables
    numeric_var_names = [key for key in dict(data.dtypes)
                         if dict(data.dtypes)[key] in ['float64', 'int64', 'float32', 'int32']]
    #get categorical variables
    cat_var_names = [key for key in dict(data.dtypes) if dict(data.dtypes)[key] not in ['float64', 'int64', 'float32', 'int32']]
    return [numeric_var_names, cat_var_names]
def list2Str(lst):
    try:
        if isinstance(lst, list):
        # print('true')
            return",".join(lst)
            # return tuple(lst)
        else:
            return lst
    except:
        print('Error in casting list: ', lst, sys.exc_info()[0])
```

```python
def process_local_data(path,Limit_Features,n_files):
    lst_df = []
    lst_files =random.sample(files, n_files)
    count = 0
    #feat = pd.read_excel(Limit_Features)
    print("Iteration Over Files, And Limitting with Imp features")
    for i in lst_files:
        try:
            temp = pd.read_json(path+i,lines = True)
#             for i in feat.Cols:
#                 if i not in temp.columns:
#                     temp[i] = np.nan
#             temp = temp[feat.Cols]
            temp = temp.loc[:,~temp.columns.duplicated()]
            cat = getNumericalAndCategorical(temp)[1]
            temp[cat] = temp[cat].applymap(lambda x:list2Str(x))
            temp[cat] = temp[cat].astype(str)
            temp = temp.fillna(np.nan)
            idx = temp[temp.event_type == 'nan'].index
            temp.drop(idx,inplace=True)
            count+=1
            if count%500 == 0:
                print(count)
            lst_df.append(temp)
        except:
            print('problmatic file name:', i)
            print("Unexpected error: for ", sys.exc_info()[0])
    print('________Merging DataFrames___________')
    return pd.concat(lst_df,ignore_index= True)
    del lst_df,lst_files,feat,count,idx,cat
    gc.collect()
```

```python
%%time
df = process_local_data('/s3data/','/home/rajatsharma/DSA/lcols.xlsx',1000)
```

```python
to_drop = ['id','type','host','path','timestamp','@timestamp','@version']+[i for i in df.columns if 'layer.eth' in i]+ [i for i in df.columns if 'layer.json' in i]
```

```python
df = df.drop(to_drop,axis = 1)
```

```python
df.shape
```

```python
def null_values(df):
    """df - is the dataframe which needs to be parsed"""
    """For checking percentage of missing data in every columns"""
    missing_val = df.isnull().sum()
    missing_perc = df.isnull().sum()*100/len(df)
    missing_df = pd.concat([missing_val, missing_perc], axis=1)
    missing_df = missing_df.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'})
    missing_df = missing_df[
        missing_df.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False).round(1)
    print ("Dataframe has " + str(df.shape[1]) + " columns.\n" 
           "There are " + str(missing_df.shape[0]) +
           " columns that have missing values.")
    return missing_df

miss_values = null_values(df)
```

```python
df = df.drop((miss_values[miss_values['% of Total Values']==100.0].index),axis = 1)
```

```python
data = df[miss_values[miss_values['% of Total Values'] <= 60].index]
```

```python
# changing Dtypes 
df[df.select_dtypes(np.float64).columns] = df.select_dtypes(np.float64).astype(np.float32)
```

```python
data.info(verbose=False,memory_usage='deep')
```

```python
#replacing values with True and False
to_replace = ['layer.tcp.tcp.flags.ack','layer.tcp.tcp.flags.cwr', 'layer.tcp.tcp.flags.ecn',
              'layer.tcp.tcp.flags.fin','layer.tcp.tcp.flags.push','layer.tcp.tcp.flags.reset',
              'layer.tcp.tcp.flags.syn','layer.tcp.tcp.flags.urg','layer.ip.ip.flags.df', 'layer.ip.ip.flags.mf']
for i in to_replace:
    data[i].replace([1.0], True, inplace= True)
    data[i].replace([0.0], False, inplace= True)
```

```python
#Replacing All None with np.nan
def is_unique(s):
    a = s.to_numpy() 
    return (a[0] == a).all()
for i in data.columns:
    if is_unique(data[i]) == False:
        data[i].replace('None', np.nan,inplace= True)
        data[i].replace('nan', np.nan,inplace = True)
```

```python
#Changing Values in "layer.tcp.tcp.options.sack_perm" with True and False
data['layer.tcp.tcp.options.sack_perm'].fillna(False,inplace= True)
data['layer.tcp.tcp.options.sack_perm'].replace('04:02', True, inplace= True)
```
